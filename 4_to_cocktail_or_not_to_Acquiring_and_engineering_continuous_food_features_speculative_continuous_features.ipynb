Make a new df with just the columns needed for the _food part_.

Drop duplicate rows.

Use `groupby`, agg, `sqeeze`, and `unstack` routine to put different categories of the source column into different columns. 


```python
#df1_limited = df[['food_id','orig_food_common_name','orig_food_part','proportion_food_part']]
df1_limited = df[['orig_food_common_name','orig_food_part','proportion_food_part']]
spread1_df = df1_limited.drop_duplicates().groupby(['orig_food_common_name','orig_food_part']).mean().squeeze().unstack().add_suffix('_proportion')
#spread1_df = df1_limited.drop_duplicates().groupby(['food_id','orig_food_part']).mean().squeeze().unstack().add_suffix('_proportion')

spread1_df.head(50)
```

Make a new df with just the columns needed for the _proportion_food part_.

Drop duplicate rows.

Use `groupby`, agg, `sqeeze`, and `unstack` routine to put different categories of the source column into different columns. 

```python
#df2_limited = df[['food_id','orig_food_common_name','orig_source_name','proportion_source']]
df2_limited = df[['orig_food_common_name','orig_source_name','proportion_source']]
#spread2_df = df2_limited.drop_duplicates().groupby(['food_id','orig_source_name']).mean().squeeze().unstack().add_suffix('_proportion')
spread2_df = df2_limited.drop_duplicates().groupby(['orig_food_common_name','orig_source_name']).mean().squeeze().unstack().add_suffix('_proportion')

#spread2_df.iloc[:,35:55]
spread2_df
```

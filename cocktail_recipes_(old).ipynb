```python
import json

def cocktail_list_from_cocktail_db():
    f=alcohol_api_url
    data = requests.get(f)    
    parsed_json = json.loads(data.text)
    drink_dict_list = []
    
   # print(parsed_json)
    for i in (parsed_json["drinks"]):
        try:
            #url = i["strDrinkThumb"]
            drink_id = i["idDrink"]
            drink_url = drink_url_stem + drink_id
                       
            r_data = requests.get(drink_url)
            drink_dict = r_data.json()['drinks'][0]
                        
#            json_url_stream = urlopen(drink_url)
#            drink_data = json.loads(json_url_stream.read())
            #print(drink_dict)
            #drink_dict = pd.read_json(drink_url)
#            drink_dict = json.loads(drink_url)
            #print(type(drink_dict))
            drink_dict_list.append(drink_dict, inplace=True)
            
            drink_dict = {i: 'string_' + str(i)}
            drink_dict_list.append(drink_dict)
        
            print(drink_dict_list)
        
            #webbrowser.open(url)
            #print(i, "\n")
            #print(i["strDrink"], "\n")
            #print(drink_url)
        except Exception:
            pass
    #print(len(tt))
    return( drink_dict_list )

cocktail_db_list = cocktail_list_from_cocktail_db()
print(cocktail_db_list)

with open("data/" + "cocktail_db_list.pickle", 'wb') as to_write:
    pickle.dump(cocktail_db_list, to_write)
#print( cocktail_db_list)
```
